```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Iris.csv")
df.head()
df.info()
```

```python
df.hist(figsize=(10,8))
plt.show()
```

```python
Histograms show the distribution of numerical features. Sepal and petal measurements are normally distributed with slight skewness.
```

```python
sns.countplot(x='species', data=df)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df.drop('species', axis=1))
plt.show()
```

```python
corr = df.drop('species', axis=1).corr()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap of Iris Features")
plt.show()
```

```python
The correlation heatmap shows a strong positive correlation between petal length and petal width. Sepal features show weaker correlations, indicating petal measurements are more influential.
```

```python
The Iris dataset is clean and contains no missing values.

The dataset is balanced across all three species: Setosa, Versicolor, and Virginica.

Petal length and petal width show clear separation between species.

Sepal features show overlapping distributions and are less discriminative.

Strong positive correlation exists between petal length and petal width.

Mild outliers are observed in some sepal measurements.

Petal features are the most important variables for species classification.
```
